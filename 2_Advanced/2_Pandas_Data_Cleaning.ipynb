```python
# Importing the libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# loading the data
df = load_dataset('lukebarousse/data_jobs')['train'].to_pandas()

# cleaning data
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

## I want to replace the na values in `salary_year_avg` with the median salary.. 
- preferably based on the `job_title_short`

```python
df.loc[:,'salary_rate':'salary_hour_avg']
```

```python
df.loc[:,'salary_rate':'salary_hour_avg'].dropna(subset= 'salary_rate')
```

```python
median_year_salary = df.salary_year_avg.median()
median_hour_salary = df.salary_hour_avg.median()
```

```python
df['salary_year_avg'].notna() #= df.salary_year_avg.median()
```

```python
filled_df = df
```

`fillna()` to fill na values in a dataframe with a specific value

```python
filled_df['salary_year_avg'].fillna(median_year_salary) 
# looking at the raplaced column but didn't assign it yet
```

```python
filled_df['salary_year_avg'] = filled_df['salary_year_avg'].fillna(median_year_salary) 
filled_df['salary_hour_avg'] = filled_df['salary_hour_avg'].fillna(median_hour_salary) 
# This will replace na values in the two columns with the median salary hourly, yearly 
```
