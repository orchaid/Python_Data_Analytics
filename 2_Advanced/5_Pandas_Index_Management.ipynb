```python
# Importing the libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# loading the data
df = load_dataset('lukebarousse/data_jobs')['train'].to_pandas()

# cleaning data
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
# to see the index attribute
df.index
```

```python
# to assign a name to the index
df.index.name = ''
```

```python
# to know what type is the index
df.index.dtype
```

# Three Methods To Cover
- `.reset_index()`
- `.set_index()`
- `.sort_index()`

### `reset_index()`

```python
df_usa = df[df.job_country == 'United States'].copy()
df_usa.head()
```

```python
df_usa.index.name = 'job_id'
df_usa.head()
```

```python
df_usa.reset_index().head(3)
# I can either drop the job_id column or keep it and use it through merging the data frame
```

### - `.set_index()`

```python
df_usa.head()
```

```python
df_usa.set_index('job_id').head(3)
# it drops the previous index and put only the new one
```

### `.sort_index()`

```python
median_pivot = df_usa.pivot_table(values= 'salary_year_avg',index= 'job_title_short',aggfunc=['median','min','max'])
median_pivot
```

```python
# I sorted the values using a tuple as there is 2 in one column
median_pivot.sort_values(by= ('median','salary_year_avg'), ascending= False) # can do the same with the index but not so useful here
```
