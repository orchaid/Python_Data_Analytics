```python
# Importing the libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading the data
df = load_dataset('lukebarousse/data_jobs')['train'].to_pandas()

# cleaning data
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Question
 How are skills trending for Data Analysts on a monthly basis.  


```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
df_DA['job_posted_month'] = df_DA['job_posted_date'].dt.strftime('%b')
```

```python
df_Da_skills = df_DA.explode('job_skills')
df_Da_skills_pivot = df_Da_skills.pivot_table(index= 'job_posted_month_no', columns= 'job_skills', aggfunc='size',fill_value= 0) # fill_value puts a certian value instead of the nan
df_Da_skills_pivot
```

We need to sort the skills to be in order of count to make it easier to plot the top X.
- easier to sort from up to down (rows) but more steps to sort from left to right (columns)



```python
df_Da_skills_pivot.sum()
```

```python
# assigning the sum to new row called it total
df_Da_skills_pivot.loc['total'] = df_Da_skills_pivot.sum()
# then sorting it's value
```

```python
# I wanna see the columns in this order so 
df_Da_skills_pivot.loc['total'].index # for the row total i want its index 
```

```python
# then sorting it's value
#loc[row:,column:]  #.index is for the names of the columns that will be sorted based on the sums of each
df_pivot = df_Da_skills_pivot[df_Da_skills_pivot.loc['total'].sort_values(ascending= False).index]
df_pivot
```

To wrap my head around the previous code:
- I wanted to oreder the columns of the pivot table left to right so:
    - I created a row called total using `df_Da_skills_pivot.loc['total'] = df_Da_skills_pivot.sum()`.
    - The index of any row is the columns above it(index and columns can be switched based on pov: tilt your head and the coulmns would be indeces and vice versa), so i used it to sort it descending: I sorted this `df_Da_skills_pivot.loc['total'].index` based on sum.
    - Then drop the total row so that it doesn't mess our graph later on.
    - If I want I can use months names instead of months numbers by creating a new column from month numbers after reseting the index and setting it a new.

```python
df_pivot.iloc[:,:5].plot(kind= 'line') # only want the trend of top 5 skills so used iloc(can limit indeces and columns) not head(limits row) for that

# didn't drop the total 
```

```python
# dropped the total 
df_pivot_no_total = df_pivot.drop(index= 'total')

df_pivot_no_total.iloc[:,:5].plot(kind= 'line')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
# If i want to see the mon as names not numbers
df_month_name = df_pivot_no_total.reset_index()
df_month_name['month'] = pd.to_datetime(df_month_name['job_posted_month_no'],format = '%m').dt.strftime('%b')
```

```python
no_month_no_df = df_month_name.drop(columns='job_posted_month_no').set_index('month').iloc[:,:5] # dropped the job_posted_month_no and assigned the index as month
no_month_no_df
```

```python
no_month_no_df.plot(kind= 'line')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
