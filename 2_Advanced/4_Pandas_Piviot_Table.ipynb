```python
# Importing the libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# loading the data
df = load_dataset('lukebarousse/data_jobs')['train'].to_pandas()

# cleaning data
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

# Counts of Job Title
- `size` counts all rows in a groub (values/columns not required)
- `count` counts non-na values in each group (values/columns required)

```python
df.groupby('job_title_short')['salary_year_avg'].size()
```

# Caluculating median salary to show the difference between 
`.pivot_table()` and `.groupby()`

```python
df.pivot_table(index= 'job_title_short', values= 'salary_year_avg', aggfunc= 'median')
```

```python
df.pivot_table(index= 'job_title_short', values= 'salary_year_avg',columns= 'job_country', aggfunc= 'median')
```

Same same, until you add columns

```python
df.pivot_table(index= 'job_country', values= 'salary_year_avg',columns= 'job_title_short', aggfunc= 'median')
 # I can switch the index and the columns if I want, to make the data frame vertical the way I like
```

# Top Countries With The Most Salary 

```python
# what are the countries with highest number of postings
df.job_country.value_counts().head(6)
```

```python
# make a list of top 6 countries
top_countries = df.job_country.value_counts().head(6).index
```

```python
# filtering based on the 6 countries
df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median' ).loc[top_countries] # loc for index filtering
```

```python
top_table = df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median' ).loc[top_countries]
```

```python
# Top 3 job postings 
df.groupby('job_title_short')['job_title'].count().sort_values(ascending=False).head(3)
```

```python
top_jobs = df.groupby('job_title_short')['job_title'].count().sort_values(ascending=False).head(3).index
top_jobs
```

```python
# filter top_table for list of 3 job titles
top_table = top_table[top_jobs] # like any column filtering
top_table
```

```python
top_table.plot(kind= 'bar')
plt.title('Median Salary by Country and Job Title')
plt.xlabel('')
plt.ylabel('median yearly salary ($USD)')
plt.xticks(rotation= 45, ha= 'right')
plt.show()
```
