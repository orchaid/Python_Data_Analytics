```python
# Importing the libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# loading the data
df = load_dataset('lukebarousse/data_jobs')['train'].to_pandas()

# Cleaning the data
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

### `iloc[]` and `loc[]` difference
- `iloc[]` works with numeric indices.
- `loc[]` works with labeled indices.

```python
df.iloc[0:3, 1:3]  # iloc[] can be used like .head(3) but i can specify the coulumn i wanna see by its indices
# notice parameter 1 works like usual indices
```

```python
df.loc[0:5, 'salary_year_avg': 'salary_hour_avg']
# notice parameter 1 doesn't work like usual indices as it gave me row 5 
```

```python
df.loc[:, 'salary_rate': 'salary_hour_avg'].dropna(subset= 'salary_rate') 
# Drops the na values in the salary_rate column
```
