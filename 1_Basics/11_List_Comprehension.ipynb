# List Comprehension
- The syntax for it `Expression for item in iterable`


```python
numbers = []
for num in range(20):
  numbers.append(num)
```

```python
# Instead of the three codes form before, i can write a list comprehension that does the same thing
number = [x for x in range(20)]
```

```python
number
```

```python
numbe = [str(x/2) for x in range(20)] # an operation to each number
# u can use a string instead of the range and it will iterate through it too
numbe
```

```python
# Not limited to only lists, can be used with tuples, sets, dicts, strings
string_example = {x for x in 'python'}
string_example
```

# Combining lists and loops
- The syntax for it  `Expression for item in iterable if condition == True`


```python
list_jobs = ['Data Analyst', 'Data Science', 'Data Engineer','Business Data Analyst']
```

```python
analyst_list = []
for job in list_jobs:
  if 'Data Analyst' in job:
    analyst_list.append(job)
# codes from a previous lesson to add data analyst jobs into a new list
```

```python
analyst_list
```

```python
analyst_list_2 = [job for job in list_jobs if 'Data Analyst' in job]
analyst_list_2
# the same output with list comprehension using less codes
```
