# `.iloc`  (Index-based selection)
`df.iloc[row_index, column_index]` works with integers only
- Stands for integer-location based indexing.
- Uses integer indices to select rows and columns.
- Works like Python's list slicing, where you specify indices to access specific rows or columns.


```python
df.iloc[0]   # Access first row (position 0)
```

```python
df.iloc[4:6] # supports slicing data from row 4 to 5
```

```python
df.iloc[4:9, 0:3] # for the first 3 column
```

```python
# much less confusing
df[['job_title','job_title_short','job_location']].iloc[4:9] # for the first 3 column
```

# `.loc` (Label-based selection)
`df.loc[row_label, column_label]`
- Stands for label-location based indexing.
- Uses row and column labels (or names) to select data.
- Can handle boolean indexing or explicit index names.

```python
df.loc[0,'job_title']
```

# Using operators to filter values
`&`-> and , `|`-> or

```python
df['job_title_short'] == 'Data Analyst'  # i can use this to filter my df based on it
```

```python
df[df['job_title_short'] == 'Data Analyst']   # shows data frame where the job_title_short is true based on my condititon  (all are data analyst jobs)
```

```python
(df['job_title_short'] == 'Data Analyst') & (df['salary_year_avg'] > 100000)  # filter further
```

```python
df['salary_year_avg'].notna()
```

```python
# to filter further based on another column

df[(df['job_title_short'] == 'Data Analyst') & (df['salary_year_avg'].notna())]  # put each condition in () and use & instead of and operator
```
