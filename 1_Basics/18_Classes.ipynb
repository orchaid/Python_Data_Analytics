```python
class TrialList:
  def __init__(self):   # __init__ is a predefined method in python to initialize variables
    """initialize an empty list."""
    self._items = []
  def add(self, item):
    """add an item to the end of the list."""
    self._items.append(item)
  def __getitem__(self, index):
    """retrieve an item by index."""
    return self._items[index]
  def __setitem__(self, index, value):
    """set an item at a specific index to a value."""
    self._items[index] = value
  def __len__(self):
    """return the length of the list."""
    return len(self._items)
  def __repr__(self):
    """returns a string representation of the list."""
    return str(self._items)
```

```python
my_list = TrialList()  # an instance of the list <variable of the list>
```

```python
my_list.add('speed')
```

```python
#  bonus_calculator(total, base)
#         calculate the bonus rate based on the total salary and the base salary.
#         Args:
#           total (float): the total salary.
#           base (float): the base salary.
#         Returns:
#           float: the bonus rate.



#     sal_calculator(base, bonus=0.1)
#         calculate the total salary based on the salary and the bonus.
#         Args:
#           base (float): the base salary.
#           bonus (rate): the bonus rate. Default is .1.
#         Returns:
#           float: the total salary.
```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate = .1,symbol= '$'):    # ctrl shift L  if i want to select all the similar words
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total = base_salary * (1+ bonus_rate)
    self.bonus = self.total - self.base_salary


  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_total(self):
    return f'{self.symbol}{self.total:,.0f}'
  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(10000)
salary.base_salary   # attribute of a salary <variables in the class>
```
