# Exercise
Filter what job to apply for based on a list of skill

```python
my_skills = ['Python','SQL','Excel']
```

```python
job_roles = [
    {"role": "Data Analyst", "skills": ["Data Visualization", "Python", "SQL", "Excel"]},
    {"role": "Business Intelligence Analyst", "skills": ["Power bi", "Python", "SQL", "Excel","Tableau"]},
    {"role": "Data Scientist", "skills": ["Machine Learning", "Python", "R", "Deep Learning"]},
    {"role": "Data Engineer", "skills": ["ETL", "SQL", "Big Data Technologies", "Cloud Computing"]},
    {"role": "Business Analyst", "skills": ["Data Interpretation", "Communication", "SQL", "Problem-Solving"]},
    {"role": "AI Engineer", "skills": ["Deep Learning", "TensorFlow", "Neural Networks", "Python"]},
    {"role": "Machine Learning Engineer", "skills": ["Machine Learning Algorithms", "Python", "TensorFlow", "Big Data"]},
    {"role": "Database Administrator", "skills": ["Database Management", "SQL", "Performance Optimization", "Backup and Recovery"]},
    {"role": "Product Analyst", "skills": ["Market Research", "A/B Testing", "Data Analysis", "SQL"]},
    {"role": "Quantitative Analyst", "skills": ["Statistical Modeling", "Mathematics", "Python", "Finance"]},
    {"role": "Data Architect", "skills": ["Data Modeling", "Cloud Platforms", "Database Design", "Big Data"]}
] # list of a dictionary
```

# Coding
- loop through the `job_roles`
- verify all `my_skills` are inside of each `job_roles`
- if `True` (or qualified), add to qualified list of jobs

## Based in the two lists above, I want to know which roles I am qualified for based on my_skills
- trying to do it step by step

```python
# This code shows only one skill in my_skills list that matches with skills in job_roles list
for job in job_roles:

  for skill in my_skills:
    if skill in job['skills']:
      print('Matched skill: ',skill)
      print (job['skills'])
      break # so that it doesn't iterate through every skill in my_skill -> to iterate only in the second list
```

```python
# i added qualified in the first loop, then in the second loop and after the second it will go back to being true as it iterates
for job in job_roles:
  qualified = True

  for skill in my_skills:
    if skill not in job['skills']: # I add the not key word to match with the qualified
      qualified = False
      print(qualified)
      print(skill,"doesn't match in:")
      print (job['skills'])
      break # to iterate through my_skills
  if qualified:
    print(job)
```

```python
# put the qualified jobs into a list
qualified_roles = []

# This will give the jobs which my_skills matches with plus some other skills that needs to be taught
for job in job_roles:
  qualified = True

  for skill in my_skills:
    if skill not in job['skills']:
      qualified = False
      break
  if qualified:
    qualified_roles.append(job['role'])
```

```python
#  all function

# Determine which job u are qualified for
qualified_role = []

for job in job_roles:
  # Check if all required skills are in my_skills
  if all(skill in job['skills'] for skill in my_skills):
    # Add the job role to the qualified_role list
    qualified_role.append(job['role'])

qualified_role
```

```python
# Or list comprehension #
qual_role = [job['role'] for job in job_roles if all( skill in job['skills'] for skill in my_skills)]
```
