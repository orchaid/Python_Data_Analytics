# Types of functions
Here's a concise table outlining the types of functions in Python:

| **Type of Function**       | **Description**                                                                 |
|----------------------------|---------------------------------------------------------------------------------|
| **Built-in Functions**     | Predefined functions in Python (e.g., `print()`, `len()`, `type()`).            |
| **User-defined Functions** | Functions created by users using the `def` keyword.                            |
| **Lambda Functions**       | Anonymous functions defined using the `lambda` keyword for simple operations.  |
| **Standard Library Functions** | Functions from Python's standard library modules (e.g., `math.sqrt()`, `os.path()`). |
| **Third-party Functions**  | Functions provided by external libraries or frameworks (e.g., `numpy.mean()`, `pandas.DataFrame()`). |
| **Recursive Functions**    | Functions that call themselves to solve problems (e.g., calculating factorial).|
| **Generator Functions**    | Functions that use `yield` to return a sequence of values lazily.              |
| **Higher-order Functions** | Functions that take other functions as arguments or return functions (e.g., `map()`, `filter()`). |
| **Class Methods**          | Functions defined in a class, bound to the class using `@classmethod`.         |
| **Instance Methods**       | Functions defined in a class, bound to an instance (usually with `self`).      |
| **Static Methods**         | Functions in a class that do not depend on class or instance (`@staticmethod`).|

 - We will focus on the first 3 in this lesson as they are out of focus for data analytics
