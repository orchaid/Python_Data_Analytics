```python
analytics_jobs = []  # initialize a list object so we can append to it

for job in job_list:
  if 'Data Analyst' in job:
    analytics_jobs.append(job)
# this will only put <copy> the values from job_list that are 'Data Analyst' into the new list called analytics_jobs
```

# While Loops
- when condition is True continue to excute

```python
count = 0
while count < 4:
  print(count)
  count += 1
```

```python
exp = 0
index = 0

lis = list(years_exp.items())

while exp < 2:
  key, exp = lis[index] #unpacking the list to two variables
  if exp >= 5:
    print(key, exp)
  index += 1

# the reason that it doesn't loop because I start with exp = 0, but as soon as I unpack key, exp = lis[index],
# exp gets replaced with the experience value of the first person in the dictionary (Alice in this case, with exp = 5),
# and doesn't continue after that because the condition becomes false <exp is not smaller than 2>.
```

```python
exp = 0
index = 0

lis = list(years_exp.items())  # Convert dictionary items to a list of tuples
while index < len(lis): # while true Loop through the entire list
    key, exp = lis[index]  # Unpack the tuple
    if exp >= 6:
        print(key,'has',exp,'years of experience')  # Print the key and experience if exp >= 8
    index += 1 # Ensure the index iterates within bounds
```
