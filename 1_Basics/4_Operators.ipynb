# Orders of Operations
- **Parentheses (Brackets)**: Solve anything inside parentheses (or brackets) first.
- **Exponents (Orders)**: Handle exponents (e.g., powers and square roots).
- **Multiplication and Division**: Perform these from left to right as they appear.
- **Addition and Subtraction**: Perform these from left to right as they appear.

```python
min_salary = 30_000  # _ python interpreter ignores it
max_salary = 50_000

avg_salary = (min_salary + max_salary) / 2
avg_salary
```
