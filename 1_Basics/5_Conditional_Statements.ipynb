# Conditional Statements
if and elif excute a block of code, if the action is true and go to the next code if false

else excutes a block of code, if the action is false

```python
if False:
  print('wow')
```

```python
if True:
  print('wow')
```

```python
# skill required by a company
job_skill = 'SQL'
# possessed skill by applicant
applicant_skill = 'looker'
years_experience = 1

if job_skill == applicant_skill:
  print('They is qualified')
elif applicant_skill == 'python':
  print('They needs to learn python')
elif years_experience >= 2:
  print('enough experience, but no required skills')
elif applicant_skill == 'excel':
  pass
else:
  print('go learn another skill or gather experience')
```
