```python
# Install datasets library (if not already installed)
!pip install datasets
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
data = load_dataset('lukebarousse/data_jobs')
df = data['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

# Bar Charts - Counts of Job Postings


```python
job_count= df.job_title_short.value_counts()
job_count
```

```python
plt.bar(job_count.index,job_count)
plt.title('Posting by Job Tilte')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45) # not aligning properly
plt.show()
```

```python
plt.bar(job_count.index,job_count)
plt.title('Posting by Job Tilte')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation =45, ha='right') # ha= 'right' aligns the x values to the right instead of the default center

plt.show()
```
