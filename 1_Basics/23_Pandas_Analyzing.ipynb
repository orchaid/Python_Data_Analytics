# syntax for groupby method
- `df.groupby('columns to group' or a list)['columns to aggregate'].your_aggregation_method() `
- `df.groupby('columns to group')['columns to aggregate' or a list].your_aggregation_method() `
- `df.groupby('columns to group')['columns to aggregate'].agg(list of aggregation methods) `

```python
df.groupby(by = 'job_country')['salary_year_avg'].max()
```

```python
country_max = df.groupby(by = 'job_country')['salary_year_avg'].median()
```

```python
country_max['Egypt']
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.groupby(['job_title_short','job_country'])['salary_year_avg'].median()     # i can groupby multible columns by putting them in the list
```

```python
# aggregate multiple columns in a list with groupby

df.groupby('job_title_short')[['salary_year_avg','salary_hour_avg']].median()
```

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['mean','median','min','max'])
```
