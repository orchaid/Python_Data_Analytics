# Logical Operators

```python
job_work_from_home = True
job_health_insurance = True
# i want a job that has these two benefits
```

```python
job_work_from_home and job_health_insurance
```

```python
job_no_degree_mention = False

job_work_from_home and job_health_insurance and job_no_degree_mention # and operator returns false if one of the variable in question is false
```

```python
job_work_from_home and job_health_insurance or job_work_from_home # or operator returns true if one of the variable in question is true
```

```python
if job_work_from_home or job_health_insurance and not job_work_from_home:
  print ('it meets the condtions')
```

# Idenity Operators `is` `is not`

```python
sal_night_shift = 10_000
sal_morning_shift = 10_000

sal_night_shift is sal_morning_shift
```

```python
# although sal_night_shift == sal_morning_shift
# because their id is not the same
print (id(sal_night_shift))
print(id(sal_morning_shift))
```

```python
my_sal = sal_night_shift # assigned them to each other, so same id

my_sal is sal_night_shift
```
