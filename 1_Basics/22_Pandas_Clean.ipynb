# Switching a column data type from str to datetime

```python
df.job_posted_date[0] #access the first row by it's index location in the column job_posted_date
# we find that it's a string
```

```python
pd.to_datetime(df.job_posted_date)
```

```python
# UBDATE into the data frame using the pandas function called t0_datetime

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.job_posted_date    # it's updated
```

# Adding a new column

```python
# because the job_posted_date is datetime data type
# we can use new property called dt.

df.job_posted_date.dt.month   # to see the month
```

```python
df['job_posted_month'] =  df.job_posted_date.dt.month   # creating a new column for the month
```

# Removing from df
 - getting rid of the salary_hour_avg column `drop(labels= 'column_name' , axis = 1)`
 - - `labels` can be a list , `axis = 0` if for removing rows
 - and removing rows where salary_year_avg is na `dropna()`
 - - 0 or index drops rows where na exist, and subset spicify list of column

```python
df.drop(labels = 'salary_hour_avg', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.dropna(subset = ['salary_year_avg'], inplace= True)
```

```python
df.info()
```
