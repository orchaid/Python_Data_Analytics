# Using format method
.format()

```python
#  |  format(...)
#  |      S.format(*args, **kwargs) -> str
#  |
#  |      Return a formatted version of S, using substitutions from args and kwargs.
#  |      The substitutions are identified by braces ('{' and '}').

'role: {}; with required skill: {} '.format(role, skill)
```

# f-strings: Formated Strings Literals

```python
f'role: {role}; required skill: {skill}' # this the best one for me
```

# % formating (printf-style string formating)

```python
'Role: %s; Skill Required: %s' % (role,skill)
```
