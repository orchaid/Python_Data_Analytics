# Classes: The template of objects
**Class**: a template for creating objects(records)

```python
job_salary = 90000
job_location = 'United States'
job_title = 'Data Analyst'

class job_post:
  def __init__ (self,title,location,salary):
    self.title = title
    self.location = location
    self.salary = salary
```

```python
job_post(job_title, job_location, job_salary) # created an instance of this object
```

# Attributes: The variable of an object
**Attribute**: a field in a record defined by its class

```python
job_1 = job_post(job_title, job_location, job_salary)
```

```python
job_1.location # we dont use parentheses after attributes because it's variable of a class
```
